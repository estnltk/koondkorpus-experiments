```python
texts = []

with open('results/koond_words_list.txt') as in_f:
    text = []
    for word in in_f:
        text.append(word.rstrip())
        if len(text) == 100:
            texts.append(' '.join(text))
            text = []
    if text:
        texts.append(' '.join(text))

len(texts)
```

```python
from estnltk.taggers import PronounTypeTagger
from estnltk import Text

tagger = PronounTypeTagger()
tagger
```

```python
from collections import defaultdict
pronouns = defaultdict(list)
```

```python
for text in texts:
    wordst = Text(text).tag_layer()
    tagger.tag(wordst)
    words2 = wordst['pronoun_type']
    for token, form, lemma, postag, ptype in zip(words2.text, words2.form, words2.lemma, words2.partofspeech, words2.pronoun_type):   
        if postag[0]  == 'P':
            for i in range(len(form)):
                if [lemma[i], postag[i], form[i], ptype[i]] in pronouns[token]:
                    continue
                else:
                    pronouns[token].append([lemma[i], postag[i], form[i], ptype[i]])
```

```python
import json
with open('pronouns.json', 'w') as fp:
    json.dump(pronouns, fp)
```

```python
with open('pronouns.json', 'r') as fi:
    pronouns_from_json = json.load(fi)
```

```python
pron_list = []
for k, v in sorted(pronouns_from_json.items()):
    if len(v) > 1:
        for v2 in v:
            pron_list.append([k, v[0][0], v[0][1], v[0][2], v[0][3]])
    else:
        pron_list.append([k, v[0][0], v[0][1], v[0][2], v[0][3]])
```

```python
import pandas as pd
df = pd.DataFrame(pron_list,columns=['token','lemma', 'postag', 'form', 'pronoun_type' ])
df[280:300]
```
